6.3 Метод `.pivot_table()` используется для создания сводных таблиц, которые позволяют агрегировать данные и представлять их в удобном формате. Основные атрибуты метода:
+ `values`: Столбец или столбцы для агрегирования
+ `index`: Столбец или столбцы, которые будут использоваться в качестве индексов строк
+ `columns`: Столбец или столбцы, которые будут использоваться в качестве заголовков столбцов
+ `aggfunc`: Функция или список функций для агрегации

В данном задании необходимо создать сводную таблицу среднего рейтинга по main_production, поэтому из списка используются только атрибуты `values`, `index` и `aggfunc`. За значения берётся рейтинг, за индексы - страна производства и функция усреднения.

Для того, чтобы выделить 3 категории года, необходимо проанализировать перцентили, используется метод `describe()`

```python
df.describe()
```

Как можно заметить, 25% значений находятся в  2008 году или раньше, медиана находится на 2014 году, 75% не позже 2018, а оставшиеся 25% - позже 2018. Таким образом, можно выделить 3 категории:

+ Средний - между 2009 и 2018
+ Свежий - после 2018

Для категоризации будет использована функция `categorize_year`

```python
def categorize_year (df):
    if df['release_year']<=2008:
        return 'Старый'
    if (df['release_year'] >2008 )& (df['release_year'] <=2018):
        return 'Средний'
    if df['release_year']>2018:
        return 'Свежий'
    
```

```python
df['release_year_category']=df.apply(categorize_year, axis=1)
```

```python
year_mean=df.pivot_table(values='release_year', index='main_genre', columns='release_year_category',aggfunc='mean' )

display(year_mean)
```

Аналогично создается сводная таблица `year_median`, содержащая медианный год по жанру и категории года

```python
year_median=df.pivot_table(values='release_year',index='main_genre',columns='release_year_category',aggfunc='median')

display(year_median)
```

```python
df.describe()
```

Как можно заметить, 25% значений с рейтингом до 7.1 медиана находится на 7.4 году, 75% не более 7.8, а оставшиеся 25% - больше 7.8. Таким образом, можно выделить 3 категории:

**Стоит обратить внимание, что категории выделяются только на основе датасета, так как в дейтсвительности оценка 7.1 - не низкая**

+ Низкий - ниже 7.1
+ Средний - между 7.1 и 7.8

Для категоризации будет использована функция `ReturnScore`

```python
def ReturnScore(df):
    if (df['Scored'])<=7.1:
        return 'Низкий'
    
    if (df['Scored']>=7.1) & (df['Scored']<=7.8):
        return 'средний'
    if df['Scored']>=7.8:
        return 'высокий'

```

Затем необходимо создать столбец Scored

```python
df['Scored']=df['score']
```

Далее с помощью метода `apply` необходимо применить функцию ко всем строкам данного столбца, для чего еще используется аргумент `axis=1`

```python
df['Scored']=df.apply(ReturnScore, axis=1)
```

Далее нужно посчитать ТОП-5 жанров фильмов с наибольшей популярностью, для этого использован метод `value_counts()`

Как можно заметить, лидируют drama, horror, comedy, crime. Documentry и romance имееют одинаковое количество записей, поэтому оба жанра попадают в топ-5

Далее необходимо отфильтровать данные с помощью метода `query`. Выбираются записи с высоким и средним рейтингом, года произведения между 2000 и 2025 и входящие в топ-5 жанров, определнных выше

```python
df = df.query('(Scored == "средний" or Scored == "высокий") and '
              'release_year > 2000 and release_year < 2025 and '
              'main_genre in ["drama", "horror", "comedy", "crime", "documentry"]')
```

Далее создаются сводные таблицы с помощью метода `pivot_table`: средняя, максимальная, минимальная, медианная по категории рейтинга и жанра

```python
df_mean=df.pivot_table(values="score",index="Scored",columns="release_year", aggfunc="mean")

df_max=df.pivot_table(values="score",index="Scored",columns="release_year", aggfunc="max")

df_min=df.pivot_table(values="score",index="Scored",columns="release_year", aggfunc="min")

df_median=df.pivot_table(values="score",index="Scored",columns="release_year", aggfunc="median")
```
