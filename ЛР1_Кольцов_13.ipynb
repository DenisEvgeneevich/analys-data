Как можно заметить, в столбце `main_production` повторов нет, а в столбце `main_genre` есть повторы:

### Задание 1

_Группировка - жанр и количество фильмов каждого `main_production`_

Группировка происходит с помощью метода `groupby()`, необходимо сгруппировать данные со столбцам `main_genre` и `main_production`, вывести количество записей, соответствующее стране производства и жанру, выведя только столбец `index` c методом `count()`

```python
df_group=df.groupby(['main_genre','main_production'])
group=df_group['index'].count()
display(group)
```

Получилась таблица, которая хранит в себе количество фильмов для каждого жанра и страны производства

### Задание 2

_Группировка - жанр и количество фильмов каждого `main_production`.
Создать датафрейм. Переименовать столбец с количеством в `“сount”`.
Отсортировать по убыванию столбца `“count”`_

Так как группировка точно такая же, то можно работать с таблицей из прошлого задания. Но теперь необходимо из этой таблицы создать датафрейм с помощью метода `pd.DataFrame()`. ### Задание 3
_Сводная таблица (pivot_table) - Сводная таблица (pivot_table) - средний рейтинг (score) фильмов по
main_production. Отсортировать по убыванию рейтинга. Округлить до одного знака_


Метод `.pivot_table()` используется для создания сводных таблиц, которые позволяют агрегировать данные и представлять их в удобном формате. Основные атрибуты метода:
+ `values`: Столбец или столбцы для агрегирования
+ `index`: Столбец или столбцы, которые будут использоваться в качестве индексов строк
+ `columns`: Столбец или столбцы, которые будут использоваться в качестве заголовков столбцов
+ `aggfunc`: Функция или список функций для агрегации

В данном задании необходимо создать сводную таблицу среднего рейтинга по main_production, поэтому из списка используются только атрибуты `values`, `index` и `aggfunc`. За значения берётся рейтинг, за индексы - страна производства и функция усреднения.
Также значения были отсортированы по убыванию и округлены до 1 знака с помощью метода `.round(1)`

```python
pivot=df.pivot_table(values='score',index='main_production',aggfunc='mean')

pivot.sort_values(by='score',ascending=False).round(1)
```

Получена таблица, показывающая средний рейтинг фильмов по каждой стране

### Задание 4
_Сводная таблица (pivot_table) - медианный рейтинг (score) фильмов по
жанрам - столбцы и main_production- строки. Отсортировать по возрастанию
main_production._

В данном задании необходимо повторить прошлую сводную таблицу, но добавить еще столбцы main_genre, сортировка проводится по `main_production` по возрастанию

```python
pivot_score=df.pivot_table(values='score',columns='main_genre',index='main_production',aggfunc='median')

display(pivot_score.sort_values(by='main_production',ascending=True).round(1))
```

Данная таблица показывает средний значения рейтинга фильмов по разным странам и жанрам,  не все страны снимают все жанры фильмов, поэтому некоторые значения заменены на `NaN`
